## Import Libraries and Load Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# The dataset
data = pd.read_csv('energy_data.csv')

# The first few rows of the dataset
data.head()
```

## Data Cleaning and Preparation

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Fill missing values with zero
data.fillna(0)
```

```python
data.info()
```

```python
# Display summary statistics
data.describe()
```

## Descriptive Statistics and Aggregation

```python
# Group by country and calculate average electricity access for the last 20 countries
avg_electricity_access1 = data.groupby('Country')['Access To Electricity (% Of Population)'].mean().reset_index().tail(20)
print(avg_electricity_access1)

# Plot the average electricity access by country for the last 20 countries
plt.figure(figsize=(12, 5))
g = avg_electricity_access1.groupby('Country', as_index=False)['Access To Electricity (% Of Population)'].sum().sort_values(by='Access To Electricity (% Of Population)', ascending=False).tail(20)
sns.barplot(x='Access To Electricity (% Of Population)', y='Country', data=g.sort_values(by='Access To Electricity (% Of Population)', ascending=False))
plt.title('Average Electricity Access by Country')
plt.xlabel('Average Electricity Access (%)')
plt.ylabel('Country')

# Print as PDF
plt.savefig('Average Electricity Access by Country.jpg')
```

```python
# Group by Country and Calculate Average Electricity access for the LEAST 20 countries
avg_electricity_access2 = data.groupby('Country')['Access To Electricity (% Of Population)'].mean().reset_index().nsmallest(20, ['Access To Electricity (% Of Population)'])
print(avg_electricity_access2)

# Plot the 20 Countries with LEAST average electricity access
plt.figure(figsize=(12, 6))
g = avg_electricity_access2.groupby('Country', as_index=False)['Access To Electricity (% Of Population)'].sum().sort_values(by='Access To Electricity (% Of Population)', ascending=False)
sns.barplot(x='Access To Electricity (% Of Population)', y='Country', data=g.sort_values(by='Access To Electricity (% Of Population)', ascending=False))                                                                                           
plt.title('Countries with Least Average Electricity Access')
plt.xlabel('Average Electricity Access (%)')
plt.ylabel('Country')

# Print as PDF
plt.savefig('Countries with Least Average Electricity Access.jpg')
```

## Time Series Analysis

```python
import warnings
warnings.simplefilter( action="ignore", category=FutureWarning)

# Average Electricity Access by Year
avg_access_by_year = data.groupby('Year')['Access To Electricity (% Of Population)'].mean().reset_index()

# Plot average electricity access over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Access To Electricity (% Of Population)', data=avg_access_by_year)
plt.title('Average Electricity Access Over Time')
plt.xlabel('Year')
plt.ylabel('Average Electricity Access (%)')

plt.savefig('Average Electricity Over Time.jpg')
```

## Energy Efficiency Optimization

```python
# Countries with high energy intensity
high_intensity = data[data['Energy Intensity Level of Primary Energy (MJ/$2017= PPP GDP)'] > data['Energy Intensity Level of Primary Energy (MJ/$2017= PPP GDP)'].mean()]
g = high_intensity.groupby('Country', as_index=False)['Energy Intensity Level of Primary Energy (MJ/$2017= PPP GDP)'].sum().sort_values(by='Energy Intensity Level of Primary Energy (MJ/$2017= PPP GDP)', ascending=False).head()

# Display the result
print(high_intensity[['Country', 'Year', 'Energy Intensity Level of Primary Energy (MJ/$2017= PPP GDP)']])

# Plot the countries with high energy intensity
plt.figure(figsize=(12, 6))
sns.barplot(x='Energy Intensity Level of Primary Energy (MJ/$2017= PPP GDP)', y='Country', data=g.sort_values(by='Energy Intensity Level of Primary Energy (MJ/$2017= PPP GDP)', ascending=False))
plt.title('Top 5 Countries with High Energy Intensity')
plt.xlabel('Energy Intensity')
plt.ylabel('Country')

# Print as PDF
plt.savefig('Top 5 Countries with High Energy Intensity.jpg')
```

```python
# Countries with High Level_of_Carbon_Emissions
Level_of_Carbon_Emissions = data[data['Carbon Emissions kt by Country'] > data['Carbon Emissions kt by Country'].mean()]
g = Level_of_Carbon_Emissions.groupby('Country', as_index=False)['Carbon Emissions kt by Country'].sum().sort_values(by='Carbon Emissions kt by Country', ascending=False).head()

print(Level_of_Carbon_Emissions[['Country', 'Year', 'Carbon Emissions kt by Country']].sort_values(by='Carbon Emissions kt by Country', ascending=False))

# Plot the countries with high energy intensity
plt.figure(figsize=(12, 6))
sns.barplot(x='Carbon Emissions kt by Country', y='Country', data=g.sort_values(by='Carbon Emissions kt by Country', ascending=False))
plt.title('Top 5 Countries with High Carbon Emissions')
plt.xlabel('Kiloton (metric) of Carbon Emissions')
plt.ylabel('Country')

plt.savefig('Top 5 Countries with High Carbon Emissions.jpg')
```

```python
# Identify regions with high renewable energy potential
high_renewable = data[data['Renewable Energy Share in the Total Final Energy Consumption (%)'] > data['Renewable Energy Share in the Total Final Energy Consumption (%)'].mean()].nlargest(20, ['Renewable Energy Share in the Total Final Energy Consumption (%)'])
r = high_renewable.groupby('Country', as_index=False)['Renewable Energy Share in the Total Final Energy Consumption (%)'].sum().sort_values(by='Renewable Energy Share in the Total Final Energy Consumption (%)', ascending=False)

# Display the result
print(high_renewable[['Country', 'Year', 'Renewable Energy Share in the Total Final Energy Consumption (%)']])

# Plot the countries with high renewable energy potential
plt.figure(figsize=(12, 6))
sns.barplot(x='Renewable Energy Share in the Total Final Energy Consumption (%)', y='Country', data=r.sort_values(by='Renewable Energy Share in the Total Final Energy Consumption (%)', ascending=False).head(50))
plt.title('Top 5 Countries with High Renewable Energy Potential')
plt.xlabel('Renewable Energy Share in the Total Final Energy Consumption(%)')
plt.ylabel('Country')

plt.savefig('Top 5 Countries with Renewable Energy Potential.jpg')
```
